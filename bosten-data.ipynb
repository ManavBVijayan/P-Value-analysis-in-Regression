```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data=pd.read_csv("housing.csv",names=columns)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
corr=data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(data=corr,annot=True)
```

```python
X=data.drop(["MEDV"],axis=1)
y=data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("R2 Score : ",r2_score(y_test,y_pred))
```

```python
# Fit linear regression with statsmodels on the standardized data
import statsmodels.api as sm
```

```python
# Add a constant term to the independent variables matrix
X_train_with_const = sm.add_constant(X_train)
```

```python
model_statsmodels = sm.OLS(y_train, X_train_with_const).fit()
```

```python
# Print the summary which contains p-values
print(model_statsmodels.summary())
```
